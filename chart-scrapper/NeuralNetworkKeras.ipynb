```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(modified_df, test_size=0.2, stratify=modified_df['position'])

train_y = train[['position']]
test_y = test[['position']]

del train['position']
del test['position']
```

```python
from sklearn.neural_network import MLPClassifier

mlp_classifier = MLPClassifier(solver = "sgd",activation = "logistic", hidden_layer_sizes=(256), verbose=True, max_iter=10000, alpha=0.0001, tol=1e-4, learning_rate_init=1)

mlp_classifier.fit(train,train_y.values.ravel())
```

```python
training_set_score = mlp_classifier.score(train,train_y.values.ravel())

training_set_score

```

```python
test_set_score = mlp_classifier.score(test,test_y.values.ravel())
test_set_score
```

```python
from sklearn.naive_bayes import GaussianNB 
gnb = GaussianNB() 
gnb.fit(train.values, train_y.values.ravel()) 
```

```python
y_pred = gnb.predict(test.values) 
```

```python
from sklearn import metrics 
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(test_y.values.ravel(), y_pred))
```
