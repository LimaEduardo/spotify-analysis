```python
import numpy as np
import pandas as pd
from pymongo import MongoClient
```

Mongo Connections

```python
client = MongoClient('localhost', 27017)
db = client['spotify_analysis']
```

Get dataset from database and convert to Pandas Dataframe

```python
dataset = db.dataset.find({})
dataframe =  pd.DataFrame(list(dataset))
```

```python
dataframe
```

Delete unuseful columns

```python
modified_df = dataframe
del modified_df["_id"]
del modified_df["music_name"]
del modified_df["artist"]

modified_df
```

Divide train and test dataset


```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(modified_df, test_size=0.2, stratify=modified_df['position'])
```

```python
train_y = train[['position']]
test_y = test[['position']]

del train['position']
del test['position']
```

```python
from sklearn.neural_network import MLPClassifier
import time

configurations = [
    (64, 64, 64, 64, 64, 64, 64, 64),
    (32, 32, 32, 32),
    (32, 32, 32, 32, 32, 32, 32, 32),
    (256, 256),
    (32, 32, 32, 64, 64, 32, 32, 32),
]

solvers = [
    "sgd"
]

activation_functions = [
    "identity", "logistic", "tanh", "relu"
]

for solver in solvers:
    
    for af in activation_functions:

        for configuration in configurations:
            
            start = time.time()

            
            print("Começou: " + solver + " / " + af + " / " + str(configuration))

            mlp_classifier = MLPClassifier(solver = solver,activation = af, hidden_layer_sizes=configuration, verbose=True, max_iter=10000, alpha=0.0001, tol=1e-4)
            

            mlp_classifier.fit(train,train_y.values.ravel())

            training_set_score = mlp_classifier.score(train,train_y.values.ravel())

            test_set_score = mlp_classifier.score(test,test_y.values.ravel())

            filename = solver + "_" + af + "_" + str(configuration).replace(', ','_').replace('(','').replace(')','')
            
            end = time.time()
            
            time_elapsed = end - start

            with open("./logs/"+filename, 'a') as out:
                file_content = "solver="+solver+"\n"
                file_content += "activation="+af+"\n"
                file_content += "layers=" + str(configuration) + "\n"
                file_content += "last_loss=" + str(mlp_classifier.loss_) + "\n"
                file_content += "last_iteration=" + str(mlp_classifier.n_iter_) + "\n"
                file_content += "training_score=" + str(training_set_score) + "\n"
                file_content += "test_score=" + str(test_set_score) + "\n"
                file_content += "time_elapsed=" + str(time_elapsed) + "\n"
                
                out.write(file_content)
                
            print("Terminou: " + solver + " / " + af + " / " + str(configuration))
            
            

```
