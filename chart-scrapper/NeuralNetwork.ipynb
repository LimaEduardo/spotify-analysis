```python
import numpy as np
import pandas as pd
from pymongo import MongoClient
```

Mongo Connections

```python
client = MongoClient('localhost', 27017)
db = client['spotify_analysis']
```

Get dataset from database and convert to Pandas Dataframe

```python
dataset = db.dataset.find({})
dataframe =  pd.DataFrame(list(dataset))
```

```python
dataframe
```

Delete unuseful columns

```python
modified_df = dataframe
del modified_df["_id"]
del modified_df["music_name"]
del modified_df["artist"]

modified_df
```

Divide train and test dataset


```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(modified_df, test_size=0.2)
```

```python
train_y = train[['position']]
test_y = test[['position']]

del train['position']
del test['position']
```

```python
from sklearn.neural_network import MLPClassifier


mlp_classifier = MLPClassifier(solver = "adam", hidden_layer_sizes=(64,64,64,64), verbose=True, max_iter=200, alpha=0.0001, tol=1e-4)

mlp_classifier.fit(train,train_y.values.ravel())

```

```python
training_set_score = mlp_classifier.score(train,train_y.values.ravel())
training_set_score
```

```python
training_set_score = mlp_classifier.score(test,test_y.values.ravel())
training_set_score
```
